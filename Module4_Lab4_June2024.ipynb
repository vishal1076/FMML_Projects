The topic of classifier in today's lab, SVMs make for really good linear separators. Let us look at an example which has linearly separable data:

```python
np.random.seed(0)

ar = np.vstack(     [\
                    np.random.multivariate_normal(np.array([1, 1]), 1.5 * np.array([[2, -1], [-1, 2.0]]), size = 50, ),\
                    np.random.multivariate_normal(np.array([3, 3]), 2 * np.array([[0.75, -0.5], [-0.5, 0.75]]), size = 50, )
                    ]\
              )

testAr = np.vstack(   [\
                      np.random.multivariate_normal(np.array([1, 1]), np.array([[0.5, -0.25], [-0.25, 0.5]]), size = 500, ),\
                      np.random.multivariate_normal(np.array([3, 3]), np.array([[0.75, -0.5], [-0.5, 0.75]]), size = 500, )
                      ]\
                  )
testy = np.array([0] * int((testAr.shape[0]/2)) + [1] * int((testAr.shape[0]/2)))

X = ar
y = np.array([0] * int((ar.shape[0]/2)) + [1] * int((ar.shape[0]/2)))

def plotDecisionBoundary(X, y, pair, clf):
  x_min, x_max = X[:, pair[0]].min() - 1, X[:, pair[0]].max() + 1
  y_min, y_max = X[:, pair[1]].min() - 1, X[:, pair[1]].max() + 1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                      np.arange(y_min, y_max, 0.1))

  y_pred = clf.predict(np.c_[xx.ravel(), yy.ravel()])
  y_pred = y_pred.reshape(xx.shape)
  plt.figure(figsize=(5,4))
  plt.contourf(xx, yy, y_pred, alpha=0.4)
  plt.scatter(X[:, pair[0]], X[:, pair[1]], c = y, s = 50, edgecolor='k')
  # plt.legend()
  plt.gcf().set_dpi(130)
  plt.show()

def boundaryExp() :
  clf = svm.LinearSVC()
  pair = [0, 1]
  clf.fit(X[:, pair], y)
  plotDecisionBoundary(X, y, pair, clf)

  plt.show()

boundaryExp()
```

## Intuitive introduction to SVMs:

![SVM](http://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1526288453/index3_souoaz.png)

SVMs are an iterative approach of trying to find the optimal hyperplane that divides the multidimentional space into different classes present in the dataset.


**Hyperplanes:** These are the decision planes that separate the objects of classes that we are trying to classify.

**Support Vectors** : Support vectors are the points from dataset that are closest to the hyperplane that divides the dataset.

**Margin**:  The gap between the closest support vectors from the different class along the direction perpendicular to the hyperplane. Simply put, it is the sum of perpendicular distance of the support vector of each class to the hyperplane.

# The Kernel Trick

The true potential of SVMs is unleashed when they are combined with kernels.

## Kernels : An intuitive explanation

Kernel methods are essentially counting on using the training data (say $i^{th}$ example $(x_i, y_i)$ ) itself in a more straightforward way and learning a corresponding weight ($w_i$) for that example. Rather than trying to learn a fixed set of parameters which is done typically.
Depending on the kind of kernel used, we can virtually project the training data in a higher dimension to make it easier for the classifier to classify them.


```python
from sklearn.datasets import make_circles
X, y = make_circles(200, factor=.1, noise=.1)

clf = svm.SVC(kernel='linear').fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='tab20')
plt.gcf().set_dpi(110)
```

```python
plt.style.use("default")
clf = svm.LinearSVC()
pair = [0, 1]
clf.fit(X[:, pair], y)
plotDecisionBoundary(X, y, [0, 1], clf)
```

However, if we artificially add another dimention to the dataset of the form:

$z = x^2 + y^2$
we can clearly see a hyperplane that can distinguish both the datasets:

```python
Z = np.array([[i[0]**2 + i[1]**2] for i in X])
```

```python
X_new = np.hstack((X, Z))
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data = [go.Scatter3d(
    x = X_new[:, 0],
    y = X_new[:, 1],
    z = X_new[:, 2],
    mode = "markers",
    marker = {
        "color" : y,
        "line": {"width" : 4, "color":'DarkSlateGrey'},
        "colorscale": "viridis"},
)])

fig.show()
```

That simple trick has helped us to get another dimension in which the data is linearly separable by a hyperplane (in this case, a 2d plane)

## **Additional Critical Thinking Questions**

Consider	a	2-class	classification	problem	in	a	2-dimensional	feature	space	x=[x1,	x2]	with	target	variable	y=1.	The	training	data	comprises	7
samples	as	shown	in	figure (4	black	diamonds	for	the	positive	class	and	3	white	diamonds	for	the	negative	class). The	maximum	margin	classifier
(also	called	linear	hard	margin	SVM)	is	a	classifier	that	leaves	the	largest	possible	margin	on	either	side	of	the	decision	boundary.	The	samples	lying	on	the	margin	are	called	support	vectors.

2. What	is	the	training	error	rate?

3. The	removal	of	which	sample	will	change	the	decision	boundary?

4. What	is	the	leave-one-out	error	rate?


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Define the training data
X = np.array([
    [1, 1],
    [2, 2],
    [2, 0],
    [3, 3],  # Positive class
    [0, 0],
    [1, 0],
    [0, 1],  # Negative class
])
y = np.array([1, 1, 1, 1, -1, -1, -1])  # Target values: 1

# Create and fit the SVM (hard margin SVM with linear kernel)
clf = SVC(kernel='linear', C=1e10)  # Large C to enforce hard margin
clf.fit(X, y)

# Extract the decision boundary and margins
w = clf.coef_[0]  # Coefficients of the decision boundary
b = clf.intercept_[0]  # Intercept of the decision boundary
support_vectors = clf.support_vectors_  # Support vectors

# Decision boundary: w[0]*x1 + w[1]*x2 + b = 0
# Margins: w[0]*x1 + w[1]*x2 + b = 1

# Generate grid for plotting
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))

# Compute decision boundary and margins
Z = w[0] * xx + w[1] * yy + b
margin_up = Z - 1
margin_down = Z + 1

# Plot data points
plt.scatter(X[y == 1, 0], X[y == 1, 1], c='black', marker='D', label='Positive Class (+1)')
plt.scatter(X[y == -1, 0], X[y == -1, 1], c='white', edgecolors='k', marker='D', label='Negative Class (-1)')
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], s=100, facecolors='none', edgecolors='red', label='Support Vectors')

# Plot decision boundary
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='k')
# Plot margins
plt.contour(xx, yy, margin_up, levels=[0], linestyles='dashed', colors='k')
plt.contour(xx, yy, margin_down, levels=[0], linestyles='dashed', colors='k')

plt.legend()
plt.title('Hard Margin SVM with Decision Boundary and Margins')
plt.xlabel('Feature x1')
plt.ylabel('Feature x2')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Training error rate
y_pred = clf.predict(X)
training_error_rate = np.mean(y_pred != y)
print(f"Training Error Rate: {training_error_rate:.2f}")

# Leave-one-out error rate
leave_one_out_errors = []
for i in range(len(X)):
    X_lo = np.delete(X, i, axis=0)
    y_lo = np.delete(y, i)
    clf_lo = SVC(kernel='linear', C=1e10)
    clf_lo.fit(X_lo, y_lo)
    y_pred_lo = clf_lo.predict([X[i]])
    leave_one_out_errors.append(y_pred_lo[0] != y[i])

loo_error_rate = np.mean(leave_one_out_errors)
print(f"Leave-One-Out Error Rate: {loo_error_rate:.2f}")
```

2. Training Error Rate

The hard margin SVM ensures a perfect separation of classes in the feature space:

Since the training data is perfectly separable, the training error rate is 0 (all points are correctly classified).
3. Impact of Removing a Sample

If a sample not on the margin is removed, the decision boundary does not change because it does not influence the margin.
If a support vector is removed, the decision boundary will change because the margin is determined by the support vectors.

4. Leave-One-Out Error Rate

The leave-one-out error rate measures how often a classifier misclassifies a sample when that sample is excluded from the training set:

If the removed sample is not a support vector, the classifier remains unchanged, so the error rate is
0
%
0% for that sample.
If a support vector is removed, the decision boundary changes, and theres a potential for misclassification. However, for separable data, the classifier can typically adapt to maintain 0 errors.
In this case, the leave-one-out error rate remains 0 because the data is linearly separable, even with the removal of any single sample.
